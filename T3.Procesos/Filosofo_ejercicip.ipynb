```python
import multiprocessing as mp
import time
import os
import random
import sys
```

```python

#--------------Fisofos----------------------

class filosofo(mp.Process):
    #  1==libre, 0==ocupado, inicializamos la lista en 1s porque todos los tenedores están libres
    def __init__(self,numero):
        '''
         0<= numero <=4
        '''
        super(filosofo, self).__init__()
        self.numero = numero
        sys.stdout.write('\nCreando al filósofo %s'%numero)
        self.tenedorIzquierdo = tenedores[numero]
        self.tenedorDerecho = tenedores[(numero+1)%8]



    def agarrarTenedor(self,lado):
        while True:
            if lado  == "izquierda":
                if self.tenedorIzquierdo == 1:
                    self.tenedorIzquierdo = 0
                    return
                else:
                    print("Tenedor izquierdo esta ocupado")
                    os.wait()

            elif lado  == "derecha":
                if self.tenedorDerecho == 1:
                    self.tenedorDerecho = 0
                    return
                else:
                    print("Tenedor derecho esta ocupado")
                    os.wait()
                    
    def dejarTenedor(self,lado):
        while True:
            if lado  == "izquierda":
                if self.tenedorIzquierdo == 0:
                    self.tenedorIzquierdo = 1
                    return
                else:
                    print("Tenedor izquierdo esta desocupado")
                    os.wait()

            elif lado  == "derecha":
                if self.tenedorDerecho == 0:
                    self.tenedorDerecho = 1
                    return
                else:
                    print("Tenedor derecho esta desocupado")
                    os.wait()

    def comer(self):
        time.sleep(1)



def notHungry():
    return (random.randint(1, 10) != 2)
    
```

```python


def rutina(filosofo,mutex):
    pname = filosofo.numero
    while(notHungry()):
        time.sleep(1)

    if (pname%2 == 1):
        try:
            mutex.acquire() #             Entramos en sección crítica, 
            filosofo.agarrarTenedor('izquierda')#     Secciones críticas
            filosofo.agarrarTenedor('derecha')
            mutex.release()
            
            sys.stdout.write('\nFilosofo %s comiendo' % filosofo.numero)
            filosofo.comer()
            
            mutex.acquire()
            filosofo.dejarTenedor('derecha')
            filosofo.dejarTenedor('izquierda')
            mutex.release()
            
        except Exception as e:
            print(e)
            os.wait()
            rutina(filosofo,mutex)
    else:
        try:
            mutex.acquire()
            filosofo.agarrarTenedor('derecha')
            filosofo.agarrarTenedor('izquierda')
            mutex.release()
            
            sys.stdout.write('\nFilosofo %s comiendo' % filosofo.numero)
            filosofo.comer()
            
            mutex.acquire()
            filosofo.dejarTenedor('derecha')
            filosofo.dejarTenedor('izquierda')
            mutex.release()
            
        except Exception as e:
            print(e)
            os.wait()
            rutina(filosofo,mutex)




def hacerfilosofo(numero,mutex):
    filosofo_actual = filosofo(numero)
    rutina(filosofo_actual,mutex)

    return


tenedores = [1]*8


filosofos = {}
for numero in range(8):
    mutex = mp.Lock()
    filosofos[numero] = mp.Process(target = hacerfilosofo, args = (numero,mutex))

for numero in range(8):
    filosofos[numero].start()
for numero in range(8):
    filosofos[numero].join()
print('\n\nTodos han comido')
```
